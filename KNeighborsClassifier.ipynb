```python
import pandas as pd
import sklearn as sk
from sklearn.datasets import fetch_openml
base=fetch_openml('mnist_784')
```

```python
base.data.shape
```

```python
base.target
```

```python
base.data["target"]=base.target
base.data
```

```python
dataset=base.data.sample(n=5000,replace=True)
dataset
```

```python
from sklearn.model_selection import train_test_split
def split():
    x=dataset.iloc[:,:-1]
    y=dataset['target']
    return train_test_split(x,y,train_size=0.8)
X_train,X_test,Y_train,Y_test=split()
```

```python
from sklearn.neighbors import KNeighborsClassifier
def predict(X_train,Y_train,k):
    neigh=KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train,Y_train)
    return neigh.predict(X_test)
Y_pred=predict(X_train,Y_train,1)
```

```python
import numpy as np
np.mean(Y_pred!=Y_test)
```

```python
ERRORS=np.array([])
for k in range(1,16):
    Y_pred=predict(X_train,Y_train,k)
    e=np.mean(Y_pred!=Y_test)
    ERRORS=np.append(ERRORS,e)
print(ERRORS)
```

```python
import matplotlib.pyplot as plt
k=range(1,16)
plt.plot(k,ERRORS)
plt.show()
```

```python
k_optimale=np.where(ERRORS==min(ERRORS))[0][0]+1
neigh=KNeighborsClassifier(n_neighbors=k_optimale)
neigh.fit(X_train,Y_train)
```

```python
predicted=neigh.predict(X_test)
```

```python

images = X_test.values.reshape((-1, 28, 28))

select = np.random.randint(images.shape[0], size=12)

fig,ax = plt.subplots(3,4)
for index, value in enumerate(select):
    plt.subplot(3,4,index+1)
    plt.axis('off')
    plt.imshow(images[value],cmap=plt.cm.gray_r,interpolation="nearest")
    plt.title('Predicted: {}'.format( predicted[value]) )
plt.show()
```
